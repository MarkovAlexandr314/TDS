```python
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import numpy as np
```

# Пункт б)

```python
# Функция нашей системы дифференциальных уравнений
def lorenz(dot,  r, b, sigma):
    x, y, z = dot
    x_der = sigma * (y - x)
    y_der = r * x - y - x * z 
    z_der = x * y - b * z
    return np.array([x_der, y_der, z_der])

# Функция получения следующий точки в соответствии с методом Рунге - Кутты
def GetDot(h, num_steps, dot0, r = 28, b = 8 / 3, sigma = 10):
    dots = [dot0]
    for i in range(num_steps):
        k1 = h * lorenz(dots[i], r, b, sigma)
        k2 = h * lorenz(dots[i] + h / 2 * k1, r, b, sigma)
        k3 = h * lorenz(dots[i] + h / 2 * k2, r, b, sigma)
        k4 = h * lorenz(dots[i] + h * k3, r, b, sigma) 
        dots.append(dots[i] + h / 6 * (k1 + 2 * k2 + 2 * k3 + k4))
    return dots

# Стартовые значения
t0 = 0
tn = 1000
num_steps = 10000
dt = (tn - t0) / num_steps
dot0 = [1, 1, 1]

# Получаем результат
result = np.array(GetDot(dt, num_steps, dot0))
x, y, z = result.T 

# Выводим график в 3d
ax = plt.figure().add_subplot(projection='3d')
ax.plot(*result.T, lw=0.5)
ax.set_title("Трёхмерный график системы Лоренца для стандартных параметров")
```
